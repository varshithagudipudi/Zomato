**Cleaning rate column**

```python
df['rate'].unique()
```

In data of rate column we have some strings like new, nan, and some symbols like -, /5 i dont want them so im just removing them to reduce the confusion.

```python
def handlerate(value):
    if(value=='NEW' or value=='-'): # if the values are NEW or - replace them with nan
        return np.nan
    else: #if they are not that values then 
        value= str(value).split('/')
        value= value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlerate)
df['rate'].head()
```

There are some null values in rate column fill that null column with mean values

```python
df.rate.isnull().sum() #Total 10019 null vlues are there in rate column
```

**Filling the null values in rate column  with mean values**

```python
df['rate'].fillna(df['rate'].mean(), inplace= True) #now the complete null values of this rate column is replaced with mean value 
df['rate'].isnull().sum()
```

```python
df.info()
```

**Droping  null values**

```python
df.dropna(inplace = True)
df.head()
```

replacing the column names of approx_cost(for two people) and listed_in(type) as cost2plates and Type

```python
df.rename(columns ={'approx_cost(for two people)':'cost2paltes', 'listed_in(type)':'Type'}, inplace= True)
df.head()
```

```python
df['location'].unique() #checking the unique location values in location column
```

there are so many unique locations are there in location column so search in listed in (city) column

```python
df['listed_in(city)'].unique()
```

compared to location column listed_in(city) column have less unique values
but these both column share the same information so let's remove one column

**removing the listed_in(city) cloumn**

```python
df = df.drop(['listed_in(city)'],axis=1)
```

let's see the unique values in cost2paltes

```python
df['cost2paltes'].unique()
```

remove the comma from every value of the cost2plates cause it is restricting  value to become an integer or a float now it is in the form of string

**Remove comma(,) from cost2paltes**

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)

df['cost2paltes']= df['cost2paltes'].apply(handlecomma)
df['cost2paltes'].unique()
```

now the values are converted into float datatype.
The cost2paltes and rate columns are cleaned

```python
df.head()
```

**Cleaning rest_type cloumn**

```python
df['rest_type'].value_counts()
```

there are some types of restaurants are really very less so let's combine those restaurants into one type

```python
rest_types = df['rest_type'].value_counts(ascending = False)
rest_types
```

```python
rest_types_lessthan1000 = rest_types[rest_types<1000] #now we can see there restaurant types which have less than 1000 restarants
rest_types_lessthan1000
```

**Making rest types less than 1000 in frequency as others**

```python
def handle_rest_type(value):
    if(value in rest_types_lessthan1000):
        return 'others'
    else:
        return value

df['rest_type']=df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

Now the data is cleaned in the rest_type. we jus merged the different type restaurants which are having less number of restaurants like less than 1000 into a one type called others 

**Cleaning the location column**

```python
df['location'].value_counts()
```

same here too in some locations there are only few restaurants let's cluster them

```python
location = df['location'].value_counts(ascending = False)
location_lessthan300 = location[location<300]

def handle_location(value):
    if(value in location_lessthan300):
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

Here, locations which are having less than 300 restaurants merge them into others
Now this column is also cleaned

```python
df.head()
```

**Cleaning cuisines column**

```python
cuisines = df['cuisines'].value_counts(ascending = False)
cuisines_lessthan100 = cuisines[cuisines<100]


def handle_cuisines(value):
    if(value in cuisines_lessthan100):
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```

cuisines which are having less than 100 values merge them together

```python
df.head()
```

The last column Type

```python
df['Type'].value_counts()
```

In Type we have less amount of unique values and good amount of restaurants so don't need change anthing.
All the columns are cleaned total data cleaning is completed
